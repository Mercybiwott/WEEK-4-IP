DATA ANALYSIS

```python
#answering questions
#identify the most popular hour of the day for picking up blue car in the city of paris
bluecar counter = df.groupby (["bluecar counter"])['hour'].sum()
bluecar counter.sort_values(ascending= False)[0:1]
```

```python
#which is the most popular hour for returning cars
cars = df.groupby(["cars"]) ['hour'].sum()
cars.sort_values(ascending =False)[0:1]
```

```python
#which station is the most popular?
STATION = df.groupby(["STATION"]).sum()
STATION.sort_values(ascending =False)
```

```python
#what postal code is the most popular for picking up bluecars
postal code = df.groupby(["postal code"])['blue cars'].sum()
postal code.sort_values(ascending= False)[0:1]
```
